# Diagramme | Plots der Accuracy und Loss

```python
folder_path = 'diagram'
loaded_images = []
numbers = [16, 32, 64, 128, 256, 512]

for j in numbers:
    img_path = os.path.join(folder_path, f'number_neurons_{j}.png')
    if os.path.exists(img_path):
        loaded_image = plt.imread(img_path)
        loaded_images.append(loaded_image)
    else:
        print(f"Bild {img_path} nicht gefunden")


num_images = len(loaded_images)
plt.figure(figsize=(50, num_images * 10))

for i, image in enumerate(loaded_images, start=1):
    plt.subplot(num_images, 1, i)
    plt.imshow(image)
    plt.axis('off')

plt.show()
```

In den ersten 5 Plots wird die höchste Accuracy ungefähr bei Epoch 6 erreicht, danach flacht der Anstieg stark ab. </br>
Der Unterschied in den ersten 5 Plots ist wie sich die Accuracy davor verändert. Außerdem verschiebt sich das erreichen der höchsten Accuracy minimal nach vorne, je höher die Neuronenzahl ist. </br>
Im letzten Plot wird schon in Epoch 4 eine höhere Accuracy erreicht, danach steigt sie nur noch langsam an. </br>

Insgesamt scheint es als würde bei einer höheren Anzahl an Neuronen die Zahl des Epoch geringer werden, in der die höchste Accuracy erreicht wird.

# Confusion Matrix

```python
folder_path = 'matrix'
loaded_images = []
numbers = [16, 32, 64, 128, 256, 512]

for j in numbers:
    img_path = os.path.join(folder_path, f'number_neurons_{j}_matrix.png')
    if os.path.exists(img_path):
        loaded_image = plt.imread(img_path)
        loaded_images.append(loaded_image)
    else:
        print(f"Bild {img_path} nicht gefunden")


num_images = len(loaded_images)
plt.figure(figsize=(60, num_images * 12))

for i, image in enumerate(loaded_images, start=1):
    plt.subplot(num_images, 1, i)
    plt.imshow(image)
    plt.axis('off')

plt.show()
```
